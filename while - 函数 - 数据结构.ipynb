#  while循环
 - 一个循环语句
 - 表示当，某条件成立的时候，就循环
 - 不知道具体循环次数，但能确定循环的成立条件的时候用while循环
 - while语法：
     
          while 条件表达式:
              语句块
 - 另外一种表达方式
          
          while 条件表达式:
              语句块1
          else:
              语句块2
              

```python
#如果说年利率是6.7%，本利是每年翻本，则多少年本钱会翻倍
benqian = 100000
year = 0
while benqian < 200000:
    benqian = benqian * (1+0.067)
    year += 1
    print("第{0}年拿了{1}块钱".format(year,benqian))
```

```python
#如果说年利率是6.7%，本利是每年翻本，则多少年本钱会翻倍
#如果拿到钱翻倍，则庆贺一下
benqian = 100000
year = 0
while benqian < 200000:
    benqian = benqian * (1+0.067)
    year += 1
    print("第{0}年拿了{1}块钱".format(year,benqian))
else:
    print("真开心！")
```

### 函数的参数和返回值
  - 参数：负责给函数传递一些必要的数据或者信息
     - 形参（形式参数）：在函数定义的时候用到的参数没有具体值，只是一个占位符。成为形参
     - 实参（实际参数）：在调用函数的时候输入的值
  - 返回值：函数的执行结果
     - 使用return关键字
     - 如果没有return，默认反回None
     - 函数一旦执行return语句，则无条件反回，即结束函数的执行

```python
#参数的定义与使用
#参数只是一个符号，代表着调用的数据
def hello(person):
    print("{0},你好!".format(person))
    print("拜拜，我走了")
    
p="明月"
hello(p)
```

## 关键字参数
  - 语法
          
              def func(p1=v1,P2=V2....):
                  func_body
                  
               调用函数：
               func（p1=value1,p2=value2.....）
  - 优点
    - 不容易混淆，一般普通实参和形参只是按照位置一一对应即可，容易出错
    - 使用关键字参数，可以不用考虑参数位置

```python
#关键字参数案例
def stu(name="No name",age=0,addr="No addr"):
    print("I an a student")
    print("我叫{0}，今年{1}岁，我住在{2}".format(name,age,addr))
    
n = "小米"
a = 18
addr= "家"
#不考虑参数位置    
stu(age=a ,name=n, addr=addr)
```

## 收集函数
  - 把没有位置，不能和定义时的参数位置相对应的参数，放入一个特定的数据结构中
  - 语法
  
           def func(*args):
               func_bady
               按照list使用方式访问args得到传入的参数
               
           调用：
           func(p1,p2,p3,p4.....)
   - 参数名args不是必须这样写，但是，我们推荐直接使用args，约定俗成
   - 参数名args前面需要有*
   - 收集参数可以和其他参数共存
   
  

```python
#收集参数代码
#函数模拟一个学生进行自我介绍，但具体内容不清楚
#srgs把他看做一个list列表

def stu( *args):
    print("hello 大家好，我自我介绍一下，简单说两句")
#type函数作用是检测变量的类型
    print(type(args))
    for item in args:
        print(item)

stu("xioaming",18,"北京","wangxioajing","single")

stu("周大神")


    
    
```

### 收集参数之关键字收集参数
  - 把关键字参数按字典格式存入收集参数
  - 语法
  
         def func( **kwargs):
             func_bady
         #调用
         func（P1=v1，p2=v2....）
  - kaargs一般约定俗成
  - 调用的时候，把多余的关键字参数放入kwargs
  - 访问kwargs需要按字典格式访问


```python
#收集参数案例

def stu ( **kwargs):
    print("hello 大家好，我自我介绍一下，简单说两句")
    print(type(kwargs))
    #对于字典的使用，python2和python3有区别
    for k,v in kwargs.items():
        print(k,"---",v)

stu(name="liyun",age=14,addr="北京")
#打印时字符串的乘法
print("*" * 20)

stu(name="周大神")
```

### 收集参数混合调用的顺序问题
  - 收集参数，关键字参数，普通参数可以混合使用
  - 使用规则就是，普通参数和关键字参数优先
  - 使用规则就是，普通参数和关键字参数优先
  - 定义的时候一般找普通参数，关键字参数，收集参数tuple，收集参数dict

```python
#收集参数混合调用案例
#stu负责模拟一个学生的自我介绍
def stu(name,age,*args,hobby="没有",**kwargs):
        print("Hello 大家好")
        print("我叫{0}，我今年{1}岁了".format(name,age))
        if hobby == "没有":
            print("没有爱好")
        else:
            print("我的爱好是{0}".format(hobby))
        print("*"*20)
        
        for i in args:
            print(i)
        print("#"*30)
        
        for k,v in kwargs.items():
            print(k,"---",v)
```

# 变量作用域
  - 变量由作用范围限制
  - 分类:按照作用域分类
    - 全局(global):在函数外部定义
    - 局部(local):在函数内部定义
  - 变量的作用范围
    - 全局变量:在整个全局范围都有效
    - 全局变量在局部可是使用(即函数内部可以使用函数外部定义的变量)
    - 局部变量在局部范围可以使用
    - 局部变量在全局范围无法使用
  - LEGB原则
    - L（Local）局部作用域
    - E（Enclosing function locale）外部嵌套函数作用域
    - G（Global module）函数定义所在模块作用域
    - B（Buildin）：python内置魔抗的作用域

# 递归函数
  - 函数直接或间接调用自身
  - 优点：简介，；理解容易
  - 缺点：对递归深度有限制，消耗资源大
  - python对递归深度有限制，超过限制报错
  - 在写递归程序的时候，一定注意结果条件

```python
#递归调用深度限制代码
x = 0
def fun():
    #申明成全局
    global x
    x += 1
    print(x)
    #函数调用自己
    fun()
#调用函数
fun()
```

```python
#斐波那契数列
#一列数字，第一个值是1，第二个也是1，从三个开始，每个数字的值等于前两个数字出现的值的和
#数学公式为：f(1) = 1 ,f(2) = 1,f(n) = f(n-1) + f(n-2)
#例如：1,1,2,3,5,8,13，，，，，
#n表示求第n个数子的斐波那契数列的值

def fib(n):
    if n == 1:
        return 1
    
    if n == 2:
        return 1
    return fib(n-1) + fib(n-2)

print(fib(6))

#汉洛塔
def move(n,a,b,c):
    if n == 1:
        print(a,"->",c)
    else:
        move(n-1,a,c,b)
        print(a,'->',c)
        move(n-1,b,a,c)
        
move(3,'A','B','C')
```

## 列表的操作
  - 访问
    - 使用下标操作（索引）
    - 列表的位置从0开始
  - 分片操作
    - 对列表进行任意一段的截取
    - l [:]

```python
# 下标访问列表
l =[1,2,3,4,6,6,8,9]
print(l[1])
```

```python
#分片操作
l =[1,2,3,4,5,6,8,9]
#注意截取的范围，包含左边下标，但是不包含右边下标
print(l[2:4])
print("*"*20)

#下标值可以为空，如果不写，左边下标值默认为0，右边下标值默认为最大值加一，即截取到最后一个数据
print(l[:])
print(l[2:])
print(l[:4])
print("*"*20)

#分片可以控制增长幅度，默认增长幅度为1
print(l[1:6:2])
print("*"*20)

#下标可以超出范围，超出后不再考虑多余下标内容
print(l[2:10])
print("*"*20)

#下标值，增长的幅度都可以为负数
#为负数，表明顺序是从右往左
#规定：数组最后一个数字的下标是-1
print(l)
#下面显示为空，为默认分片总是从左往右截取，即正常情况，分片值一定要小于右边的值
print(l[-1:-3])

print(l[-3:-1])
#如果分片一定要左边值比右边大，则幅长参数需要使用负数
#此案例为一个list直接正反颠倒提供了一种思路
print(l[-1:-3:-1])
print("*"*20)
```
